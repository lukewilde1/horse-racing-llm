```python
import pandas as pd
import numpy as np

print("=== Loaded Libraries ===")
```

```python
df = pd.read_csv(r"..\data\final_combined_racing_data.csv")
df.count()
```

```python
missing_counts = df.isna().sum()
print(missing_counts)
```

```python
df["Finish Position"] = df["Finish Position"].fillna(5)
df["Place"] = df["Finish Position"].apply(lambda x: 1 if x <= 3 else 0)
df["Win"] = df["Finish Position"].apply(lambda x: 1 if x == 1 else 0)
# df.drop(columns=["Finish Position"], inplace=True)
```

```python
missing_counts = df.isna().sum()
print(missing_counts)
```

```python
df['Trainer'].fillna('UNKOWN', inplace=True)
```

```python
available_classes = df["Class"].dropna().unique()
df["Class"] = df["Class"].apply(lambda x: np.random.choice(available_classes) if pd.isna(x) else x)
```

```python
df['Track Condition'].fillna('UNKOWN', inplace=True)
df['Weather'].fillna('UNKOWN', inplace=True)
df['Jockey'].fillna('UNKOWN', inplace=True)
```

```python
print(df.loc[df["SP"].isna(), ["Date", "Track", "Race Number", "Horse", "Jockey", "Trainer"]])
```

```python
# If the SP isn't there fill using per race median
df["SP_missing"] = df["SP"].isna().astype(int)

# Fill missing with race-level median
df["SP"] = df.groupby(["Date", "Track", "Race Number"])["SP"].transform(
    lambda x: x.fillna(x.median())
)

# Fallback to overall median (if entire race missing SP)
df["SP"].fillna(df["SP"].median(), inplace=True)
```

```python
filtered = df[(df["Track"] == "SHA TIN") & (df["Date"] == "2025-10-19")]
print(filtered)
```

```python
# Get Jockey Place Rate
jockey_place_rate = (
    df.groupby("Jockey")["Place"]
      .mean()
      .reset_index(name="PlaceRate")
)
jockey_place_rate["PlaceRatePct"] = jockey_place_rate["PlaceRate"] * 100

print(jockey_place_rate.sort_values("PlaceRatePct", ascending=False))
```

```python
# Get Jockey Win Rate
jockey_win_rate = (
    df.groupby("Jockey")["Win"]
      .mean()
      .reset_index(name="JockeyWinRate")
)

print(jockey_win_rate.sort_values("JockeyWinRate", ascending=False) * 100)
```

```python
# Trainer Win rates
trainer_win_rate = (
    df.groupby("Trainer")["Win"]
        .mean()
        .reset_index(name="TrainerWinRate")
    )

print(trainer_win_rate.sort_values("TrainerWinRate", ascending=False) * 100)
```

```python
# Horse win rates
horse_win_rate = (
    df.groupby("Horse")["Win"]
        .mean()
        .reset_index(name="HorseWinRate")
)

print(horse_win_rate.sort_values("HorseWinRate", ascending=False) * 100)
```

```python
# Merge the win rates as decimal
df = df.merge(jockey_win_rate[["Jockey", "JockeyWinRate"]], on="Jockey", how="left")
df = df.merge(trainer_win_rate[["Trainer", "TrainerWinRate"]], on="Trainer", how="left")
df = df.merge(horse_win_rate[["Horse", "HorseWinRate"]], on="Horse", how="left")
df
```

```python
# Encode the Trainer, Horse and Jockey
from sklearn.preprocessing import LabelEncoder

le_jockey = LabelEncoder()
le_trainer = LabelEncoder()
le_horse = LabelEncoder()

df["Jockey_ID"] = le_jockey.fit_transform(df["Jockey"])
df["Trainer_ID"] = le_trainer.fit_transform(df["Trainer"])
df["Horse_ID"] = le_horse.fit_transform(df["Horse"])

df["Jockey_Enc"] = df["JockeyWinRate"]
df["Trainer_Enc"] = df["TrainerWinRate"]
df["Horse_Enc"] = df["HorseWinRate"]

# Fill missing values
df["JockeyWinRate"].fillna(df["JockeyWinRate"].mean(), inplace=True)
df["TrainerWinRate"].fillna(df["TrainerWinRate"].mean(), inplace=True)
df["HorseWinRate"].fillna(df["HorseWinRate"].mean(), inplace=True)
```

```python
# Export to csv for future use
df.to_csv(r'features_encoded.csv', index=False)
```
